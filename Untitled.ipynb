```python
import pandas as pd
import re
```

```python
data = pd.read_csv('Data/itmo_26.04.2020_23:16.csv', index_col=0)
```

```python
data.head()
```

```python
data = data.fillna(' ')
```

```python
data['Clear_keywords'] = data['Ключевые слова ВШЭ'].apply(lambda x: re.sub('Раздел', '', x))
data['Clear_keywords'] = data['Clear_keywords'].apply(lambda x: re.sub('Тема', '', x))
data['Clear_keywords'] = data['Clear_keywords'].apply(lambda x: re.sub(r'\d+\.', '', x))
data['Clear_keywords'] = data['Clear_keywords'].apply(lambda x: re.sub(r'\d+\.\d+\.?', '', x))
data['Clear_keywords'] = data['Clear_keywords'].apply(lambda x: re.sub(r'- \d+', '', x))
```

```python
data['Clear_keywords'] = data['Clear_keywords'].apply(lambda x: '; '.join(['; '.join(key_phrase.split('.')) for key_phrase in x.split('@@@')]))
```

```python
data['Clear_keywords'] = data['Clear_keywords'].apply(lambda x: '; '.join([k_w.strip() for k_w in x.split(';') if k_w.strip()]))
```

```python
data['Keywords'] = data['Clear_keywords'] + data['Ключевые слова']
```

```python
data = data.drop(['Clear_keywords', 'Ключевые слова', ])
```
